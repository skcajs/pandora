```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
N = 100   # Number of time steps
Delta_t = 0.1  # Time step size

# System parameters (unknown, to be estimated)
k_true = 2.0  # Spring constant
b_true = 0.5  # Damping coefficient
m_true = 1.0  # Mass

# True parameter vector
theta_true = np.array([0, Delta_t, -k_true * Delta_t / m_true, -b_true * Delta_t / m_true, 0, 0])

# Initial state
x_true = np.array([[1.0], [0.0]])  # Initial position and velocity
F_k = 1.0  # Constant force applied

# Initial Estimates
x_hat = np.zeros((2, 1))  # Initial state estimate
theta_hat = np.zeros((6, 1))  # Initial parameter estimate

# Covariance matrices
P_x = 0.1 * np.eye(2)
P_theta = 10**8 * np.eye(6)
R_x = np.identity(2) # Measurement noise covariance
R_theta = np.identity(2)  # Parameter covariance

# Forgetting factors
lambda_x = 0.9
lambda_theta = 0.9985

# Cross covariance
Gamma = np.zeros((2, 6))

# Store results
x_estimates = []
theta_estimates = []
time = np.arange(N) * Delta_t

# Observer loop
for k in range(N):
    # Generate the regressor matrix Phi_k
    Phi_k = np.array([
        [x_true[0, 0], x_true[1, 0], F_k, 0, 0, 0],
        [0, 0, 0, x_true[0, 0], x_true[1, 0], F_k]
    ])
    
    # Compute observer gains
    K_x = P_x @ np.linalg.inv(P_x + R_x)
    K_theta = P_theta @ Gamma.T @ np.linalg.inv(Gamma @ P_theta @ Gamma.T + R_theta)
    Gamma = (np.identity(2) - K_x) @ Gamma
    
    # Measurement (simulate real-world observation)
    y_k = x_true  # Assume perfect measurements (no noise)
    
    # Estimate state and parameters
    innovation = y_k - x_hat
    x_hat = x_hat + (K_x + Gamma @ K_theta) @ innovation
    theta_hat = theta_hat - K_theta @ innovation
    
    # Prediction step
    x_hat = x_hat + Phi_k @ theta_hat
    
    # Update covariance matrices
    P_x = (1 / lambda_x) * (np.identity(2) - K_x) @ P_x
    P_theta = (1 / lambda_theta) * (np.identity(6) - K_theta @ Gamma) @ P_theta
    Gamma = Gamma - Phi_k
    
    # Store results
    x_estimates.append(x_hat.flatten())
    theta_estimates.append(theta_hat.flatten())
    
    # True system update
    x_true = x_true + np.array([
        [0, Delta_t],
        [-k_true * Delta_t / m_true, -b_true * Delta_t / m_true]
    ]) @ x_true + np.array([[0], [Delta_t / m_true]]) * F_k

# Convert lists to arrays
x_estimates = np.array(x_estimates)
theta_estimates = np.array(theta_estimates)

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(time, x_estimates[:, 0], label="Estimated Position")
plt.plot(time, x_estimates[:, 1], label="Estimated Velocity")
plt.xlabel("Time (s)")
plt.ylabel("State")
plt.legend()
plt.title("Adaptive Observer: State Estimation")
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(time, theta_estimates[:, 2], label="Estimated k/m")
plt.plot(time, theta_estimates[:, 3], label="Estimated b/m")
plt.axhline(-k_true / m_true, linestyle="--", color="gray", label="True k/m")
plt.axhline(-b_true / m_true, linestyle="--", color="gray", label="True b/m")
plt.xlabel("Time (s)")
plt.ylabel("Estimated Parameters")
plt.legend()
plt.title("Adaptive Observer: Parameter Estimation")
plt.show()
```
